```python
# Jared Campbell
# jsc3642
# EE 351M HW 5.1

#imports
import matplotlib.pyplot as plt
import numpy as np
import scipy.io
%matplotlib inline 
plt.style.use('ggplot')
#load data from matlab
data = scipy.io.loadmat('hw5mat.mat')
x = data['x'].T
d = data['d'].T
#init vars
m = 50
w = np.zeros([m+1,1])
u = np.zeros([m+1,1])
e = np.zeros([1000,])
e_squared = np.zeros([1000,])
#LMS
for n in range(1000):
    if(n<m+1):
        u[:n+1] = x[n+1:0:-1]
    else:
        u = x[n-m:n+1]
        u = np.flipud(u)
    y = np.dot((w.T), u)
    e[n] = d[n] - y
    w = w + 2*0.01*e[n]*u
    e_squared[n] = e[n]**2
#plot graphs
print('1. (a)')    
plt.stem(w)
plt.title('Coefficiants of Transversal Filter of Length 50')
plt.xlabel('index')
plt.ylabel('w')
plt.show()
plt.figure()
print('1.(b)')
plt.stem(e_squared[:500])
plt.title('First 500 Samples of Squared Error')
plt.xlabel('Time')
plt.ylabel('Squared Error')
plt.show()
print('''
1. (c) Finite Impulse Response because the coeffeciants die down after about 40 samples
''')
```
