```python
'''
Input: ,csv with names 
Output: .txt with contact information for each name
'''


import requests
from bs4 import BeautifulSoup as BS
import csv

with open('Book1.csv','r') as csvfile:
    data = list(csv.reader(csvfile))
d2 = []
for row in data:
    row = "".join(row)
    d2.append(row)
open('emails.txt', 'w').close()
file = open('emails.txt','w')    
    
def print_info(name):
    arr = []
    url = 'https://directory.utexas.edu/'
    payload = {'q':name}
    r = requests.get(url, params=payload)
    with open("requests_results.html", "wb") as f:
        f.write(r.content)
    html = open('requests_results.html', "r")
    soup = BS(html)
    
    elements = soup.findAll("td")
    for tag in elements:
        if tag == elements[3]:
            t1 = tag.extract()
            t1 = t1.get_text()
            t1 = t1.strip()
            arr.append(t1)
            break
        else:
            txt = tag.contents[0]
            trimmed_txt = txt.strip()
            arr.append(trimmed_txt)
    return arr

def clean(a1):
    i=0
    while i < len(a1)-1:
        a1[i] = a1[i] + " " + a1[i+1]
        a1.remove(a1[i+1])
        i+=1
    return a1


for n in d2:
    a1 = print_info(n)
    a1 = clean(a1)
    print(a1)
    for i in a1:
        file.write(i)
        file.write('\n')
    file.write('\n')
```
